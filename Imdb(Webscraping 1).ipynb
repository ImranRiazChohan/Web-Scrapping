```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
```

```python
url='https://www.imdb.com/search/title/?release_date=2017&sort=num_votes,desc&page=1'
response=get(url)
```

```python
html_soup=BeautifulSoup(response.text,'html.parser')
```

```python
movie_containers=html_soup.find_all('div',class_='lister-item mode-advanced')
print(type(movie_containers))
print(len(movie_containers))
```

```python
first_vote=first_movie.find('span',attrs={'name':'nv'})
first_vote=int(first_vote['data-value'])
first_vote
```

```python
first_metascore=first_movie.find('span',class_="metascore favorable").text
first_metascore=int(first_metascore)
first_metascore
```

```python
Names=[]
Ratings=[]
Votes=[]
Years=[]
Metascores=[]

for container in movie_containers:  
    if container.find('div', class_ = 'ratings-metascore') is not None:
# The name
        Name=container.h3.a.text
        Names.append(Name)
    
        Year=container.find('span',class_='lister-item-year').text
        Years.append(Year)
    
        Rating=float(container.strong.text)
        Ratings.append(Rating)
    
        Vote=container.find('span',attrs={'name':'nv'})
        Vote=int(Vote['data-value'])
        Votes.append(Vote)
        
        m_score = container.find('span', class_ = 'metascore').text
        Metascores.append(int(m_score))

        
```

```python
df=pd.DataFrame({'Name':Names,'Ratings':Ratings,'Vote':Votes,"MetaScore":Metascores,'Year':Year})
```
